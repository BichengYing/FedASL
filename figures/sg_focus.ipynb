```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

fedavg_data = pd.read_csv("fedavg_arbitrary_cifar10_0.002_markov_3.log")
fedasl_data = pd.read_csv("fedasl_arbitrary_cifar10_0.002_markov_3.log")
fedau_data = pd.read_csv("fedau_arbitrary_cifar10_0.002_markov_3.log")
scaffold_data = pd.read_csv("scaffold_arbitrary_cifar10_0.002_markov_3.log")
```

```python
fedasl_data.columns = ["Time", "Remove", "Iter", "Mode", "Loss", "Acc"]
fedasl_data = fedasl_data.drop("Remove", axis=1)
fedasl_data["Iter"] = fedasl_data["Iter"].map(int)
fedasl_data = fedasl_data.dropna(axis=0)

fedau_data.columns = ["Time", "Remove", "Iter", "Mode", "Loss", "Acc"]
fedau_data = fedau_data.drop("Remove", axis=1)
fedau_data["Iter"] = fedau_data["Iter"].map(int)
fedau_data = fedau_data.dropna(axis=0)

fedavg_data.columns = ["Time", "Remove", "Iter", "Mode", "Loss", "Acc"]
fedavg_data = fedavg_data.drop("Remove", axis=1)
fedavg_data["Iter"] = fedavg_data["Iter"].map(int)
fedavg_data = fedavg_data.dropna(axis=0)

scaffold_data.columns = ["Time", "Remove", "Iter", "Mode", "Loss", "Acc"]
scaffold_data = scaffold_data.drop("Remove", axis=1)
scaffold_data["Iter"] = scaffold_data["Iter"].map(int)
scaffold_data = scaffold_data.dropna(axis=0)
```

```python
def train_data(df):
    return df.query("Mode == ' train' and Iter <= 10000")[::3]


def eval_data(df):
    return df.query("Mode == ' eval' and Iter <= 10000")[::3]
```

```python
plt.figure(figsize=(5, 3.3))
plt.plot(
    eval_data(fedasl_data).Iter,
    eval_data(fedasl_data).Acc.astype(float),
    "-o",
    markevery=20,
    markersize=8,
    label="FOCUS",
)
plt.plot(
    eval_data(fedau_data).Iter,
    eval_data(fedau_data).Acc.astype(float),
    "-^",
    markevery=20,
    markersize=8,
    label="FedAU",
)
plt.plot(
    eval_data(scaffold_data).Iter,
    eval_data(scaffold_data).Acc.astype(float),
    "-s",
    markevery=20,
    markersize=8,
    label="SCAFFOLD",
)
plt.grid(True)
plt.xlabel("Communication Round")
plt.ylabel("Test Accuracy")
plt.title("Passive Arbitrary Client Participation")
plt.legend()
```
